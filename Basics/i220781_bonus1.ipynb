### Bonus Task

You are managing an inventory system for a warehouse. The inventory is represented as a list of nested lists, where each sublist contains information about an item, including its name (string), quantity in stock (integer), and price per item (float). Write a Python program to perform the following tasks: <br>
•	First, initialize the inventory list with the following data: ["Apples", 50, 1.2], ["Bananas", 30, 0.5], ["Oranges", 20, 0.8], and ["Mangoes", 10, 1.5].<br> 
•	Then, create a function named update_inventory(item_name, quantity_sold) that takes the name of an item and the quantity sold as inputs. This function should decrease the quantity of the specified item in the inventory by the quantity sold. If the item is out of stock after the update, remove it from the inventory. <br>
•	Next, create another function named generate_report() that calculates and displays the total revenue generated by selling all the items and lists all remaining items in stock along with their quantities and total value (quantity × price). <br>
•	Finally, write a loop that allows the user to repeatedly sell items by entering their name and quantity, and view the updated report after each transaction, until the user decides to quit.<br>


```python
inventory = [["Apples", 50, 1.2] , ["Bananas", 30, 0.5] , ["Oranges", 20, 0.8] , ["Mangoes", 10, 1.5]]

sold = []
```

```python
def update_inventory(item_name, quantity_sold):
    for item in inventory:
        if item[0] == item_name:
            if quantity_sold > item[1]:
                print(f"Cannot sell {quantity_sold} {item_name}(s). Only {item[1]} available.")
                return
            item[1] -= quantity_sold
            sold.append([item_name, quantity_sold, item[2]])
            if item[1] == 0:
                inventory.remove(item)
            return
    print(f"Item {item_name} not found in inventory.")
    
```

```python
def generate_report():
    total_revenue = 0
    
    for sold_item in sold:    
        total_revenue += sold_item[1] * sold_item[2]
        
    print("Total revenue: ", total_revenue)
    print("Remaining items in stock:")
    for item in inventory:
        print('Item name: ', item[0], '\tRemaining quantity: ', item[1], '\tTotal value: ', item[1] * item[2])
    return
```

```python
while True:
    item_name = input("Enter the name of the item you want to sell: ")
    quantity_sold = int(input("Enter the quantity you want to sell: "))
    update_inventory(item_name, quantity_sold)
    generate_report()
    if input("Do you want to continue selling items? (y/n): ") == 'n':
        break
```
